```
from torchvision import transforms, utils, datasets
from torch.utils import data
import matplotlib.pyplot as plt
import numpy as np
```

```
batch_size = 16
train_loader = data.DataLoader(
    datasets.CIFAR10('/content/drive/My Drive/.data',
                     train=True,
                     download=True,
                     transform=transforms.Compose([
                         transforms.ToTensor()
                     ])),
    batch_size=batch_size
)

test_loader = data.DataLoader(
    datasets.CIFAR10('/content/drive/My Drive/.data',
                     train=False,
                     download=True,
                     transform=transforms.Compose([
                         transforms.ToTensor()
                     ])),
    batch_size=batch_size
)

```

```
iterator = iter(train_loader)
images, labels = iterator.next()
images.shape
```

```
img = utils.make_grid(images, padding=0)
img.shape
npimg = img.numpy()
plt.imshow(np.transpose(npimg, (1, 2, 0)))
plt.show()
```

```
classes = {
    0: 'plane', 
    1: 'car', 
    2: 'bird', 
    3: 'cat',
    4: 'deer', 
    5: 'dog', 
    6: 'frog', 
    7: 'horse', 
    8: 'ship', 
    9: 'truck'
}

for label in labels:
  print(classes[label.item()])
```

```
idx = 0
item_img = images[idx]
item_img.shape
item_npimg = item_img.numpy()
plt.title(classes[labels[idx].item()])
plt.imshow(np.transpose(item_npimg, (1, 2, 0)))
plt.show()
```
